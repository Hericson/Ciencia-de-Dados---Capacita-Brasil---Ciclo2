```python
# Média de felicidade por sexo:

mask_alunos_feminino = alunos['sexo'] == 'F'
mask_alunos_masculino = alunos['sexo'] == 'M'
```

```python
alunos[mask_alunos_feminino]['nivel_felicidade_semanal'].mean()
```

```python
alunos[mask_alunos_masculino]['nivel_felicidade_semanal'].mean()
```

```python
horas_estudo_por_sexo = {'M': alunos[mask_alunos_masculino]['horas_estudo_semanais'].mean(),
                         'F': alunos[mask_alunos_feminino]['horas_estudo_semanais'].mean()}

pd.DataFrame.from_dict(horas_estudo_por_sexo, orient='index').plot.bar(title='Horas Estudo Semanais por Sexo')
```

```python
def plot(coluna, title):
    valores = {'M': alunos[mask_alunos_masculino][coluna].mean(),
               'F': alunos[mask_alunos_feminino][coluna].mean()}

    pd.DataFrame.from_dict(valores, orient='index').plot.bar(title=title)
```

```python
plot(coluna='horas_series_semanais', title='Atividade Física por Sexo')
```
